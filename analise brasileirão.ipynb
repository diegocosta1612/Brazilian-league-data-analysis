```python
#importação de bibliotecas

import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
#importando minha base de dados csv das últimas 22 temporadas do brasileirão

df = pd.read_csv(r"C:\Users\diego\OneDrive\Área de Trabalho\analise brasileirão\brasileirao.csv")
display(df)
```

```python
# criando novas estatísticas com base nos meus dados existentes:

df["percentual_vitorias"] = (df["won"]/df["played"])*100
df["media_gols_marcados"] = df["goals"]/df["played"]
df["media_gols_sofridos"] = df["goals_taken"]/df["played"]

display(df)
```

```python
#Análise dos campeões em busca de insights, para isso preciso criar um filtro que retorne apenas o campeão de cada temporada.
filtro_campeoes = df["place"] == 1

df_campeoes = (df[filtro_campeoes])
display(df_campeoes)
```

```python
#qual foi o maior campeão em número de campeonatos?
df_campeoes["team"].mode()

#A função moda me retornou dois valores pois os doi times possuem o mesmo número de títulos na minha base de dados
```

```python
#quantas vezes cada time foi campeão?

df_campeoes.groupby("team")["place"].count().sort_values(ascending=False)


#com a análise de dados foi possível verificar que houve uma hegemonia de times paulistas nos últimos 20 anos, somando 12 dos 20 títulos.
```

```python
#expressando a tabela acima de forma gráfica
df_campeoes_contagem = df_campeoes.groupby("team")["place"].count().sort_values(ascending=False).reset_index(name="titulos")
sns.barplot(data=df_campeoes_contagem, x="team", y="titulos")
plt.title("Campeões nos últimos 20 anos")
plt.xlabel("Time")
plt.ylabel("Número de Títulos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
