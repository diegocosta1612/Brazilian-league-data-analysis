```python
#importação de bibliotecas

import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
#importando minha base de dados csv das últimas 22 temporadas do brasileirão

df = pd.read_csv(r"C:\Users\diego\OneDrive\Área de Trabalho\analise brasileirão\brasileirao.csv")
display(df)
```

```python
# criando novas estatísticas com base nos meus dados existentes:

df["percentual_vitorias"] = (df["won"]/df["played"])*100
df["media_gols_marcados"] = df["goals"]/df["played"]
df["media_gols_sofridos"] = df["goals_taken"]/df["played"]

display(df)
```

```python
#Análise dos campeões em busca de insights, para isso preciso criar um filtro que retorne apenas o campeão de cada temporada.
filtro_campeoes = df["place"] == 1

df_campeoes = (df[filtro_campeoes])
display(df_campeoes)
```

```python
#qual foi o maior campeão em número de campeonatos?
df_campeoes["team"].mode()

#A função moda me retornou dois valores pois os doi times possuem o mesmo número de títulos na minha base de dados
```

```python
#quantas vezes cada time foi campeão?

df_campeoes.groupby("team")["place"].count().sort_values(ascending=False)


#com a análise de dados foi possível verificar que houve uma hegemonia de times paulistas nos últimos 20 anos, somando 12 dos 20 títulos.
```

```python
#expressando a tabela acima de forma gráfica
df_campeoes_contagem = df_campeoes.groupby("team")["place"].count().sort_values(ascending=False).reset_index(name="titulos")
sns.barplot(data=df_campeoes_contagem, x="team", y="titulos")
plt.title("Campeões nos últimos 20 anos")
plt.xlabel("Time")
plt.ylabel("Número de Títulos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Análise estatística dos campeões:
df_campeoes.describe()




#percebemos que os campeõestem uma taxa de vitória na casa dos 60%
#os campeões possuem um ataque muito forte pois a média de gols por partida é 1.75, ou seja, vale a pena investir na contratação de uma equipe de atacantes mais poderosa
#ao mesmo tempo que também é interessante possuir uma defesa sólida, a média de gols sofridos é menos de um gol por jogo.

#Uma conclusão interessante é que o ataque é o principal determinante para a vitória de um time no campeonato.
```

```python
#analisando a última tabela percebo que um time obteve um resultado surpreendente, superando os 70% de vitórias, irei extrair esse time agora:
indice_maior_campeao = df_campeoes["percentual_vitorias"].idxmax()
df_campeoes.loc[indice_maior_campeao]


#O time mais eficinete foi o flamengo em 2019
```

```python
#Analisando qual time possui a melhor defesa:
indice_defesa = df_campeoes["media_gols_sofridos"].idxmin()
df_campeoes.loc[indice_defesa]


#É possível observar que a melhor defesa foi o são paulo na temporada de 2007.
```

```python
sns.lmplot(data = df, y="goals", x="place", line_kws={"color":"red"})
plt.title("Correlação entre gols marcados e posição no campeonato")
```

```python
sns.lmplot(data = df, y="goals_taken", x="place", line_kws={"color":"red"})
plt.title("Correlação entre sofridos e posição no campeonato")
```

```python
cols=[
    "place",
    "goals_taken",
    "goals"
]

sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm")
plt.title("comparando a influencia entre gols sofridos e marcados na colocação final do campeonato")

#analisando o mapa de calor percebi que a colocação sofre mais influeência dos gols sofridos do que dos gols marcados.
#agora é possível perceber com clareza a importância da defesa e do goleiro no resultado final dos times.
```

```python
sns.lmplot(data = df, y="loss", x="place", line_kws={"color":"red"})
plt.title("Correlação da derrota com a posição final")

sns.lmplot(data = df, y="won", x="place", line_kws={"color":"red"})
plt.title("Correlação do vitória com a posição final")

sns.lmplot(data = df, y="draw", x="place", line_kws={"color":"red"})
plt.title("Correlação do empate com a posição final")


#já era esperado que a correlação entre derrotas e a posição final fosse tão influente, porém foi verificado que o número de empates não influenciou tanto quanto esperávamos.
```

```python
cols_campeoes=[
    "place",
    "won",
    "loss",
    "draw"
]

sns.heatmap(df[cols_campeoes].corr(), annot=True, cmap="coolwarm")
plt.title("Buscar a vitória a qualquer custo aumentando as chances de uma derrota é uma estratégia interessante?")


#percebemos que a correlação é praticamente a mesma entre a posição no campeonato e o número de vitórias ou de derrotas, logo não perder tem o mesmo peso de ganhar (é óbvio, mas mesmo assim quis tirar a prova, sou apenas um engenheiro curioso que não entende muito de futebol :( ))
```
